Before drawing a bar chart, mapping Global sales of either a platform, a Genre or a Publisher, over the years, you will need to summarise.  So if you're going for Genre, get the sum over Year and Genre.

```r
# Summarize the data by genre and year, computing the sum of global sales
genre_sales <- vgsales_clean %>% 
  group_by(Genre, Year) %>% 
  summarize(Global_Sales = sum(Global_Sales), .groups = "drop")
```

Draw a bar chart, of total sales  for a single platform / Genre / Publisher over each year.

```r
# Select the desired genre
genre <- "Action"
genre_sales_filtered <- genre_sales %>% filter(Genre == genre)

# Create a bar chart of the filtered data
ggplot(genre_sales_filtered, aes(x = Year, y = Global_Sales)) + 
  geom_col(fill = "steelblue") + 
  labs(x = "Year", y = "Global Sales", title = paste("Total Sales for Genre:", genre))
```

Draw a scatter plot (geom_point) for the same data

```r
# Select the desired genre
genre <- "Action"
genre_sales_filtered <- genre_sales %>% filter(Genre == genre)

# Create a scatter plot of the filtered data
ggplot(genre_sales_filtered, aes(x = Year, y = Global_Sales)) + 
  geom_point(color = "red") + 
  labs(x = "Year", y = "Global Sales", title = paste("Total Sales for Genre:", genre))
```

Draw a stacked bar chart, for all platforms OR genres OR publishers over each year.

```r
# Create a stacked bar chart of the genre sales over each year
ggplot(genre_sales, aes(x = Year, y = Global_Sales, fill = Genre)) + 
  geom_bar(stat = "identity") +
  scale_fill_hue() + 
  labs(x = "Year", y = "Global Sales", title = "Total Sales by Genre and Year")
```

Draw a scatter plot for the same data.

```r
# Create a scatter plot of the genre sales over each year
ggplot(genre_sales, aes(x = Year, y = Global_Sales, color = Genre)) + 
  geom_point() +
  scale_color_hue() + 
  labs(x = "Year", y = "Global Sales", title = "Total Sales by Genre and Year")
```

Draw a line chart for a single value of Platform / Publisher  / Genre.

```r
# Filter the genre_sales data for a single genre
genre_sales_single <- genre_sales %>% filter(Genre == "Sports")

# Create a line chart of the global sales over each year for the selected genre
ggplot(genre_sales_single, aes(x = Year, y = Global_Sales, group = 1)) +
  geom_line(color = "blue") +
  labs(x = "Year", y = "Global Sales", title = "Total Sales of Sports Genre by Year")
```

Draw a stacked line chart for all values of Platform / Publisher / Genre.

```r
vgsales_clean$Year <- as.integer(vgsales$Year)


genre_sales <- vgsales_clean %>%
  group_by(Genre, Year) %>%
  summarize(Global_Sales = sum(Global_Sales), .groups = "drop")

# Create stacked line chart of global sales by genre and year
ggplot(data = genre_sales, aes(x = Year, y = Global_Sales, color = Genre)) +
  geom_line(stat = "identity") +
  scale_color_brewer(palette = "Set1") +
  labs(title = "Global Sales by Genre and Year", x = "Year", y = "Global Sales", color = "Genre")
```

Draw an area chart for all values of Platform / Publisher / Genre.

```r
# Create area chart of global sales by genre and year
ggplot(data = genre_sales, aes(x = Year, y = Global_Sales, fill = Genre)) +
  geom_area() +
  labs(title = "Global Sales by Genre and Year", x = "Year", y = "Global Sales")
```

Draw a small multiples chart for all values of Platform / Publisher / , noting that depending on which side of the tilda (~) your column is, the multiples will be row-wise or column wise.

```r
# Create a small multiples chart of global sales by genre
ggplot(genre_sales, aes(x = Year, y = Global_Sales)) +
  geom_line(aes(color = Genre)) +
  labs(title = "Global Sales by Genre over Time",
       x = "Year",
       y = "Global Sales") +
  theme_bw() +
  facet_wrap(~Genre, scales = "free_y", ncol = 4)
```
