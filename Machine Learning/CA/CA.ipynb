```python
import numpy as np
import pandas as pd 

my_file = open('./data/feature_names.txt', 'r')
data = my_file.read()
data_into_list = data.split("\n")
my_file.close()

dataset=pd.read_csv('./data/dataset.csv', names=data_into_list, header= None)
dataset
```

Data Specific Continuous Features

```python
Continuous = dataset[['id', 'age', 'fnlwgt','education-num','capital-gain','capital-loss','hours-per-week']]
Continuous
```
